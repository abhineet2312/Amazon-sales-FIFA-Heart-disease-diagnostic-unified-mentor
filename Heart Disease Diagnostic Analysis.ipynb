```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Heart Disease data.csv')
```

```python
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation among all the attributes',fontsize=25)
```

```python
df['target'].value_counts()
```

```python
sns.countplot(x='target', data=df, palette='Set2')
plt.xlabel('Target', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Heart Disease Rates', fontsize=15)
plt.show()
```

```python
sns.countplot(x='sex', hue='target', data=df, palette='Set1')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Heart Disease by Gender', fontsize=15)
plt.legend(title='Target', loc='upper right')
plt.show()
```

```python
sns.histplot(x='age', hue='target', data=df, bins=20, kde=True, palette='viridis')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Heart Disease by Age', fontsize=15)
plt.legend(title='Target', loc='upper right')
plt.show()
```
